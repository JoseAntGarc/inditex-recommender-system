```python
import json 

def create_target_dict(df):
    # Asegurar que partnumber y session_id tengan los tipos correctos
    df['partnumber'] = df['partnumber'].astype(int)
    df['session_id'] = df['session_id'].astype(str)
    
    # Filtra las filas con add_to_cart_pred == 1
    positive_df = df[df['add_to_cart_pred'] == 1]
    # Filtra las filas con add_to_cart_pred == 0
    negative_df = df[df['add_to_cart_pred'] == 0]
    
    # Agrupa los partnumbers con add_to_cart_pred == 1 por session_id
    grouped_positive = positive_df.groupby('session_id')['partnumber'].apply(list)
    
    # Agrupa los partnumbers con add_to_cart_pred == 0 por session_id
    grouped_negative = negative_df.groupby('session_id')['partnumber'].apply(list)
    
    # Construye las listas de exactamente 5 elementos
    session_ids = df['session_id'].unique()
    target_dict = {}
    
    for session_id in session_ids:
        # Obt√©n los partnumbers con add_to_cart_pred == 1
        positive_parts = grouped_positive.get(session_id, [])
        
        # Si hay menos de 5 elementos, usa partnumbers con add_to_cart_pred == 0 para rellenar
        if len(positive_parts) < 5:
            negative_parts = grouped_negative.get(session_id, [])
            positive_parts.extend(negative_parts[:5 - len(positive_parts)])
        
        # Asegurar que la lista final tenga exactamente 5 elementos
        target_dict[int(session_id)] = positive_parts
    
    # Construye el diccionario final
    result = {"target": target_dict}
    return result

def export_to_json(data, filename):
    with open(filename, 'w') as f:
        json.dump(data, f, indent=4)  # `indent=4` para un formato legible
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import random 

partnumber_to_index = {pn: idx for idx, pn in enumerate(products_filtered['partnumber'])}

# Funci√≥n principal
def complete_dict(data_dict, products, product_embeddings, partnumber_to_index):
    for key, values in data_dict.items():
        # Si ya tiene 5 o m√°s valores, pasamos
        if len(values) >= 5:
            continue
        
        # Seleccionar un partnumber al azar de los valores actuales
        partnumber = random.choice(values)
        
        # Verificar si el partnumber existe en el mapeo
        if partnumber not in partnumber_to_index:
            print(f"Warning: partnumber {partnumber} not found in products DataFrame. Skipping key {key}.")
            continue
        
        # Obtener el embedding del partnumber
        product_idx = partnumber_to_index[partnumber]
        embedding_target = product_embeddings[product_idx].reshape(1, -1)
        
        # Calcular similitud del coseno con todos los embeddings
        similarities = cosine_similarity(embedding_target, product_embeddings)[0]
        
        # Ordenar los √≠ndices de mayor a menor similitud
        similar_indices = np.argsort(similarities)[::-1]
        
        # Agregar los partnumbers m√°s similares
        for idx in similar_indices:
            similar_partnumber = products.iloc[idx]['partnumber']
            if similar_partnumber not in values:
                values.append(similar_partnumber)
            if len(values) == 5:
                break
        
        # Actualizar el diccionario
        data_dict[key] = values
    
    return data_dict

```

```python
product_embeddings = np.array(products_filtered['embedding'].tolist())
final_dict = {"target": complete_dict(create_target_dict(pred_test)['target'], products_filtered,product_embeddings,partnumber_to_index)}
final_dict
```

```python
def convert_to_serializable(obj):
    if isinstance(obj, dict):
        return {k: convert_to_serializable(v) for k, v in obj.items()}
    elif isinstance(obj, list):
        return [convert_to_serializable(v) for v in obj]
    elif isinstance(obj, np.integer):  # Si es un entero de NumPy
        return int(obj)
    return obj  # Retornar el valor sin cambios si no requiere conversi√≥n

export_to_json(convert_to_serializable(data_dict),'predictions_3.json')
```
